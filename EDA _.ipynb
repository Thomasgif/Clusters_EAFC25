##Distribución del "overall":

Para observar esta distribución, creamos primero un arreglo con el rango usual del *overall* y luego se crea un histograma con la información del *overall* de los jugadores:

```python
bins_ = np.arange(40, 100, 1)
plt.figure()
sns.histplot(x = "OVR",
             data = datos, bins = bins_, kde = True, color = "slateblue")
plt.title("Distribución del overall")
plt.xlabel("Overall")
plt.ylabel("Frecuencia")
plt.grid( "--", alpha = 0.5)
plt.show()
```

##Overall por posición mediante diagrama de cajas (Boxplot):

Se observará ahora la distribución del *overall* por posición me diante un diagrama de cajas o cuartiles:

```python
posiciones = datos.groupby('Position')['OVR'].median().sort_values().index
posiciones
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(10, 6))

sns.boxplot(x = 'Position', #Nombre de la variable en el data frame
            y = 'OVR', #Nombre de la variable dependiente
            data = datos, #Conjunto de datos
            order = posiciones, #Variable creada con antelación
            palette = 'Blues_d')
plt.title('Overall rating by position')
plt.ylabel('Overall rating')
plt.show()
```

##Matriz de correlación

Luego, se creará la matriz de correlación, asignando antes los atributos en una variables:

```python
matrix = datos[['OVR', 'PAC', 'SHO', 'PAS', 'DRI','DEF', 'PHY']].corr()
maska =np.triu(np.ones_like(matrix, dtype = bool))

plt.figure(figsize = (8, 6))
sns.heatmap(matrix, annot = True, cmap = 'copper', linewidths = 0.5) #mask = maska)
plt.title('Matriz de correlación de los atributos de los jugadores')
plt.show()
```

Finalmente, observaremos los atributos de los jugadores por cada posición. Para esto, se configura la figura y los ejes para las sub-gráficas, se representa la distribución de cada atributo en un subgrupo separado y se ajusta la gráfica:

```python
import warnings
warnings.filterwarnings('ignore')

attributes = ['PAC', 'SHO', 'DRI', 'PHY', 'DEF', 'PAS']

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(18, 15), sharey=True,)
fig.suptitle('Player attributes by position', fontsize=16)

axes = axes.flatten()


for i, attr in enumerate(attributes):
    order = datos.groupby('Position')[attr].median().sort_values().index

    sns.boxplot(x='Position', y=attr, data=datos, ax=axes[i], order=order, palette = 'mako')
    axes[i].set_title(f'{attr} rating by position')
    axes[i].set_ylabel(f'{attr}')
    axes[i].set_xlabel('Position')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

#Pre-procesamiento

Del conjunto de datos, se mantienen aquellas columnas que serán relevantes para realizar el K-means, mediante lo siguiente:

```python
column_objetivo = ['Name', 'OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Acceleration',
       'Sprint Speed', 'Positioning', 'Finishing', 'Shot Power', 'Long Shots',
       'Volleys', 'Penalties', 'Vision', 'Crossing', 'Free Kick Accuracy',
       'Short Passing', 'Long Passing', 'Curve', 'Dribbling', 'Agility',
       'Balance', 'Reactions', 'Ball Control', 'Composure', 'Interceptions',
       'Heading Accuracy', 'Def Awareness', 'Standing Tackle',
       'Sliding Tackle', 'Jumping', 'Stamina', 'Strength', 'Aggression',
       'Weak foot', 'Skill moves', 'Age', 'GK Diving', 'GK Handling',
       'GK Kicking', 'GK Positioning', 'GK Reflexes', 'Preferred foot',]
column = datos.columns

for col in column:
  if col not in column_objetivo:
    datos.drop(col, axis = 1, inplace = True)
  else:
    datos[col] = datos[col]
```

###Imputación de los datos:

Ahora, revisaremos los datos faltantes o nulos (null).

```python
datos.isna().sum()
```

Luego, se imputan los datos faltantes para las estadisticas de arquero de otros jugadores, cambiándolos de NaN a 0, de la siguiente forma:

```python
gk_columns = ['GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes']
datos[gk_columns] = datos[gk_columns].fillna(0)
```

```python
x = datos.iloc[:, 1:].values
x
```

A continuación, se codifica el pie preferido (*Prefered foot*) en valores binarios (0 y 1), mediante sklearn.preprocesing y creando el codificador con LabelEncoder.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

x[:,-7] = le.fit_transform(x[:,-7])

print(x[:,-7])
```

Luego, los datos que posean (si es que los hay) serán reemplazados por 0.

```python
df = pd.DataFrame(x)
for col in df.columns:
  df[col] = pd.to_numeric(df[col], errors='coerce')

df
```

```python
X = df.values

X
```

###Imputación de datos faltantes mediante sklearn:

En este caso, los datos faltantes (si es que los hay) serán reemplazados por la mediana.

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer.fit(X)
X = imputer.transform(X)

X
```
